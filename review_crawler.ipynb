```python
# you need to install chromedriver

import selenium
from selenium import webdriver
import bs4
import lxml
import pandas as pd
import time
```

```python
# init dataframe
# sentiment gonna be caculated by specific method has standard
df = pd.DataFrame(columns=['id','title','review','score','sentiment'])
df
```

```python
# get naver movie reviews
base_url = 'https://movie.naver.com/movie/point/af/list.naver'

# execute chrome web driver
driver = webdriver.Chrome('./chromedriver.exe')
```

```python
for p in range(1,1001):
    driver.get(''.join([base_url,'?&page=',str(p)])) # get target page
    driver.implicitly_wait(time_to_wait=3) # in second
    
    html = bs4.BeautifulSoup(driver.page_source,'lxml')
    reviews:bs4.element.ResultSet = html.find('table',{'class':'list_netizen'}).find('tbody').find_all('tr')
    
    lst = []
    # get id, title, review text and score
    for r in reviews:
        ac_num = r.find('td',{'class':'ac num'}).text
        t = r.find('td',{'class':'title'})
        title = t.find('a').text
        score = float(t.find('em').text)
        
        t.a.decompose()
        t.span.decompose()
        t.div.decompose()
        t.br.decompose()
        t.a.decompose()
        
        text = t.get_text(strip=True)
        
        lst.append([ac_num,title,text,score,None])
    df = pd.concat([df,pd.DataFrame(lst,columns=['id','title','review','score','sentiment'])],ignore_index=True)

driver.close()

df.to_csv()
df
```

```python
df.to_csv(''.join(['./data/reviews_',str(int(time.time())),'.csv']),index=False)
```
